```python
import gymnasium as gym
from gymnasium.wrappers import RescaleAction
import numpy as np
import time

def reward_function(observation, action):
    diag_q = [1,10,1,1]; 
    r = 1
    #print("observation:", observation)
    #print("observation:", observation[0,1])
    cost = diag_q[0]*(observation[0]**2) + diag_q[1]*(observation[1]**2) +\
                diag_q[2]*(observation[2]**2) + diag_q[3]*(observation[3]**2) +\
                r*(action**2)

    return -cost

env = gym.make('InvertedPendulum-v4',render_mode = "human")
min_action = -20
max_action = 20
env = RescaleAction(env, min_action=min_action, max_action=max_action)
obs, _ = env.reset()
#print('obs = ', obs.dtype)

q_pos = np.array([0,np.pi/4])
q_vel = np.array([0,0])
env.set_state(q_pos, q_vel)

for i in range(10):
    action =  env.action_space.sample()
    print("action:", action)
    
    observation, reward, termination, truncated, info = env.step(action)
    reward = reward_function(observation, action)
    
    env.render()

    time.sleep(1)
    print("step:", i, "observation:", observation," reward:", reward, " termination", termination, " truncated:", truncated)
    if abs(observation[0])>= 10:
        observation, info = env.reset()
    #     env.set_state(q_pos, q_vel)

env.close()
```

```python
import gymnasium as gym
import imageio
from stable_baselines3 import DDPG
from stable_baselines3.common.monitor import Monitor

# Create the environment and wrap it with the Monitor to record video
# env = gym.make('CartPole-v1')
env = gym.make('InvertedPendulum-v4',render_mode = "human")

env = Monitor(env)

# Instantiate the PPO agent
model = DDPG('MlpPolicy', env, verbose=1)

# Train the agent
model.learn(total_timesteps=10)

# Save the trained model
model.save("ppo_cartpole")

# Load the trained model
model = DDPG.load("ppo_cartpole")

# Reset the environment to record video
obs = env.reset()

# List to store frames for the video
frames = []

# Run the agent in the environment and collect frames
for _ in range(10):
    frames.append(env.render())  # Capture the current frame
    action, _states = model.predict(obs)
    obs, reward, done, info = env.step(action)
    # frames.append(env.render())  # Capture the current frame
    if done:
        obs, info = env.reset()

# Close the environment
env.close()

# Save the frames as a video using imageio
video_path = "ppo_cartpole_video.mp4"
imageio.mimsave(video_path, frames, fps=30)

print(f"Video saved to {video_path}")
```
